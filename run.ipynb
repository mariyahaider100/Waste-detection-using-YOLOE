```python
import matplotlib.pyplot as plt
from ultralytics import YOLOE
from PIL import Image
import numpy as np
from IPython.display import display

# Initialize a YOLOE model
model = YOLOE("yoloe-11l-seg.pt")  # or select yoloe-11s/m-seg.pt for different sizes

# Set text prompt to detect the target class (example: camel)
names = ["camel"]
model.set_classes(names, model.get_text_pe(names))

# Run detection on the given image (use raw string for Windows path)
img_path = r"D:\Waste Detection\camel.webp"
results = model.predict(img_path, conf=0.25)

# Get first result
res = results[0]

# Use res.plot() to get an annotated image array or PIL image and display inline
try:
    annotated = res.plot()
    # If it's a PIL Image, display directly
    if isinstance(annotated, Image.Image):
        display(annotated)
    else:
        # assume numpy array (possibly BGR). Convert to RGB for matplotlib.
        arr = np.array(annotated)
        if arr.ndim == 3 and arr.shape[2] == 3:
            # Heuristic: convert BGR->RGB
            arr = arr[..., ::-1]
        plt.figure(figsize=(8, 8))
        plt.imshow(arr)
        plt.axis('off')
        plt.show()
except Exception as e:
    print("res.plot() failed:", e)
    # As a fallback, try to get a visualization via res.show() but avoid external viewers.
    try:
        # Some environments return a path from res.show(); capture that if possible.
        res.show()
    except Exception as e2:
        print("res.show() also failed:", e2)
```
